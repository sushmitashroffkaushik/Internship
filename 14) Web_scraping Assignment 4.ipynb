```python
# Question 1

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

driver = None
def init_driver(url):
    global driver
    try:
        options = webdriver.ChromeOptions()
        options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36")
        driver = webdriver.Chrome(options=options)
        driver.get(url)
    except Exception as e:
        print("Error:", e)
    return driver

def scrape_details():
    table = driver.find_element(By.CSS_SELECTOR, "table.wikitable")
    rows = table.find_elements(By.TAG_NAME, 'tr')[1:]
    for index, row in enumerate(rows):
        columns = row.find_elements(By.TAG_NAME, 'td')
        # Check if there are enough columns
        if len(columns) >= 5:
            rank = index + 1
            name = columns[0].text.strip()
            artist = columns[1].text.strip()
            views = columns[2].text.strip()
            upload_date = columns[3].text.strip()
            print(f"Rank: {rank}, Name: {name}, Artist: {artist}, Upload Date: {upload_date}, Views: {views} billions")
        else:
            print("Skipping row with insufficient columns.")
    return None

def main():
    try:
        init_driver("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
        if driver:
            scrape_details()
    finally:
        if driver:
            driver.quit()

if __name__ == "__main__":
    main()
```

```python
# Question 2

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = None

def init_driver(url):
    global driver
    try:
        options = webdriver.ChromeOptions()
        options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36")
        driver = webdriver.Chrome(options=options)
        driver.get(url)
    except Exception as e:
        print("Error:", e)
    return driver

def scrape_details():
    link = driver.find_element(By.CSS_SELECTOR, 'a[data-element_text="Fixtures"]')
    if link:
        link.click()
        WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div.match-card'))
        )
        fixtures = driver.find_elements(By.CSS_SELECTOR, 'div.match-card')
        for fixture in fixtures:
            series = fixture.find_element(By.CSS_SELECTOR, '.match-tournament-name').text.strip()
            place = fixture.find_element(By.CSS_SELECTOR, '.match-place').text.strip()
            date = fixture.find_element(By.CSS_SELECTOR, '.match-dates').text.strip()
            time = fixture.find_element(By.CSS_SELECTOR, '.match-time').text.strip()

            print(f"Series: {series}, Place: {place}, Date: {date}, Time: {time}")
    
def main():
    try:
        init_driver("https://www.bcci.tv/")
        scrape_details()
    finally:
        if driver:
            driver.quit()

if __name__ == "__main__":
    main()
```

```python
# Question 4

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support import expected_conditions as EC
import time

driver = None
def init_driver(url):
    global driver
    try:
        options = webdriver.ChromeOptions()
        options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36")
        driver = webdriver.Chrome(options=options)
        driver.get(url)
    except Exception as e:
        print("Error:", e)
    return driver

def navigate():
   list_root = driver.find_element(By.TAG_NAME, "ul")
   el = list_root.find_elements(By.CSS_SELECTOR, "li button")[2]
   actions = ActionChains(driver)
   actions.move_to_element(el).perform()
   list_root.find_element(By.CSS_SELECTOR, 'a[href="/trending"]').click()
   return None 
    
def scrape_details():
  rows = driver.find_elements(By.CSS_SELECTOR, "article.Box-row")
  for i in range(0, len(rows)):
    ActionChains(driver).scroll_by_amount(0, i * 50)
    time.sleep(2)
    row = driver.find_elements(By.CSS_SELECTOR, "article.Box-row")[i]
    title = row.find_element(By.TAG_NAME, 'h2').text.strip()
    desc = "No Description"
    try:
      desc = row.find_element(By.TAG_NAME, 'p').text.strip()
    except:
      desc = "No Description"
    row.find_element(By.CSS_SELECTOR, "h2 a.Link").click()
    WebDriverWait(driver, 30).until(
      EC.presence_of_element_located((By.XPATH, "//span[text()='Code']"))
    )
    count = 0
    try:
      count =  driver.find_elements(By.CSS_SELECTOR, "h2 a span.Counter")[1].text.strip() if len(driver.find_elements(By.CSS_SELECTOR, "h2 a span.Counter")) == 0 else driver.find_elements(By.CSS_SELECTOR, "h2 a span.Counter")[0].text.strip()
    except:
      count = 0
    lang = ",".join([x.text.strip() for x in driver.find_elements(By.CSS_SELECTOR, 'a[data-ga-click="Repository, language stats search click, location:repo overview"] span.text-bold')])
    print(f"Title: {title}, Description: {desc}, Count: {count}, Lang: {lang}")
    driver.back()
  return None

def main():
    try:
        init_driver("https://github.com/")
        if driver:
          navigate()
          WebDriverWait(driver, 30).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, "article.Box-row"))
          )
          scrape_details()
    finally:
        if driver:
            driver.quit()

if __name__ == "__main__":
    main()
```

```python
# Question 8

import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = None
def init_driver(url):
    global driver
    try:
        options = webdriver.ChromeOptions()
        options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36")
        driver = webdriver.Chrome(options=options)
        driver.get(url)
    except Exception as e:
        print("Error:", e)
    return driver

def scrape_details():
    link = driver.find_element(By.XPATH, "//a[@href='/datasets']")
    if link:
        link.click()
        time.sleep(5)
    rows = driver.find_elements(By.CSS_SELECTOR, 'div.rounded-box[role = "row"]')
    index = len(rows)
    for i in range(0, len(rows)):
        row = driver.find_elements(By.CSS_SELECTOR, 'div.rounded-box[role = "row"]')[i]
        row.find_element(By.CSS_SELECTOR, "h2 a").click()
        time.sleep(2)
        name = driver.find_element(By.TAG_NAME, "h1").text.strip()
        type = driver.find_element(By.XPATH, "//h1[contains(text(), 'Feature Type')]/following-sibling::p").text.strip()
        task = driver.find_element(By.XPATH, "//h1[contains(text(), 'Associated Tasks')]/following-sibling::p").text.strip()
        no_of_instances = driver.find_element(By.XPATH, "//h1[contains(text(), '# Instances')]/following-sibling::p").text.strip()
        year = driver.find_element(By.XPATH, "//h2[contains(text(), 'Donated on')]").text.strip()
        print(f"Name {name}, Type {type}, Feature Type {type}, No. of Instances {no_of_instances}, Associated Tasks {task}, Year {year}")
        driver.back()
        time.sleep(1)
    return None

def main():
    try:
        init_driver("https://archive.ics.uci.edu/")
        if driver:
            scrape_details()
    finally:
        if driver:
            driver.quit()

if __name__ == "__main__":
    main()
```
